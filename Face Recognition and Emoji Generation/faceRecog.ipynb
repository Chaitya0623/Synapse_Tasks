```python
import cv2
# pip install opencv-pythonS
# pip install opencv-contrib python
```

```python
from deepface import DeepFace
# pip install deepface
```

```python
img = cv2.imread('happyboy.jpg')
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)) #Default image is in BGR
```

```python
predictions = DeepFace.analyze(img)
```

```python
predictions['emotion']
```

```python
predictions['dominant_emotion']
```

# In Order to Draw a Rectangle

```python
faceCascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
# haarcascade_frontalface is a face recognition algorithm
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
faces = faceCascade.detectMultiScale(gray,1.1,4)

for(x,y,w,h) in faces:
    cv2.rectangle(img, (x,y), (x+w,y+h), (0,255,0), 2)
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

# In Order to Overlay a Text

```python
font = cv2.FONT_HERSHEY_SIMPLEX

# putText is used to add text to an image
cv2.putText(img,
    predictions['dominant_emotion'],
    (50, 50),
    font, 2,
    (0,255,0),
    cv2.LINE_4
) 
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```
